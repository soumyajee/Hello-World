```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine_data=pd.read_csv('winequality-red.csv')
wine_data
```

```python
wine_data = wine_data.iloc[:,:]
wine_data.head()
```

```python
wine_data
```

```python
wine_normal = scale(wine_data)
```

```python
wine_normal
```

```python
pca = PCA(n_components = 10)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="red")
```

```python
 pca_values[:,4:7]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,4:8],columns=['pc1','pc2','pc3','pc4']), wine_data[['quality']]], axis = 1)
finalDf
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import seaborn as sns
sns.scatterplot(data=finalDf,x='pc1',y='pc2',hue='quality',s = 100)
```
